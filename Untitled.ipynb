```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('presdata.csv')



df['total_state_votes'] = df.groupby(['year', 'state'])["votes"].transform('sum')

df['total_nation_votes'] = df.groupby(['year'])["votes"].transform('sum')

df['candidate_state_votes'] = df.groupby(['year', 'state', 'party'])["votes"].transform('sum')

df['candidate_nation_votes'] = df.groupby(['year', 'party'])["votes"].transform('sum')

df['percentage_of_county_votes'] = (df['votes'] / df['total_county_votes']) * 100

df['percentage_of_state_votes'] = (df.groupby(['year', 'state', 'party'])["votes"].transform('sum') / df['total_state_votes']) * 100

df['percentage_of_nation_votes'] = (df['candidate_nation_votes'] / df['total_nation_votes']) * 100

df['county_as_percentage_of_state_votes'] = (df['total_county_votes'] / df['total_state_votes']) * 100

df['county_as_percentage_of_nation_votes'] = (df['total_county_votes'] / df['total_nation_votes']) * 100

df['state_as_percentage_of_nation_votes'] = (df['total_state_votes'] / df['total_nation_votes']) * 100

filter2 = df["state"].isin(["Ohio"])
filter3 = df["county"].isin(["Medina"])
filter4 = df[("party")].isin(["democrat", "republican"])

df[filter2 & filter3 & filter4].pivot_table(values="percentage_of_county_votes", index="year", columns="party").plot()

df[filter2 & filter3 & filter4].pivot_table(values="state_as_percentage_of_nation_votes", index="year",).plot()

df.sort_values(by=['year'])[filter2 & filter3 & filter4]
```
